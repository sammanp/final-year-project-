```python
# Build a Markov Chain Model
def build_markov_chain(text_list, n=2):
    markov_chain = defaultdict(list)

    for text in text_list:
        words = text.split()
        for i in range(len(words) - n):
            key = tuple(words[i:i + n])
            markov_chain[key].append(words[i + n])

    return markov_chain

# Function to generate recipe directions
def generate_recipe(markov_model, start_words=("chicken", "curry"), length=50):
    current_words = start_words
    generated_text = list(current_words)

    for _ in range(length):
        next_word_choices = markov_model.get(current_words, None)
        if not next_word_choices:
            break  # Stop if no next word

        next_word = random.choice(next_word_choices)
        generated_text.append(next_word)
        current_words = (current_words[1], next_word)  # Move to next key

    return " ".join(generated_text)
```
